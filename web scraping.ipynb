```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd

 
#filter url or normal url
url = input('Enter the URL of Internshala \n')
#example URL
#url="https://internshala.com/internships/backend%20development%2Cfront%20end%20development%2Cfull%20stack%20development%2Cweb%20development-internship/ppo-true"


next = True

#salary for filter
sal_must = int(input('Enter Least Salary : '))


i=0
internship_detail = {}
while next:
    print(url)
    response = requests.get(url)
    response
    data = response.text
    soup = BeautifulSoup(data,'html.parser')
    intern_box = soup.find_all('div',{'class':'individual_internship'})

    
    
    for intern in intern_box:
        title = intern.find('a').text
        link = "https://internshala.com" + intern.find('a').get('href')
        company = intern.find('a',{"class":"link_display_like_text"}).text
        location = intern.find('a',{"class":"location_link"}).text
        duration = intern.find_all("td")[1].text
        salary = intern.find_all("td")[2].text
        
        
        
        #salary find
        matchob = re.match(r'(.*)-(.*) /Month (.*)',salary,re.M|re.I)
        if matchob:
            sal = int(matchob.group(2))
        else:
            matcho = re.match(r'(.*) /Month (.*)',salary,re.M|re.I)
            if matcho:
                sal = int(matcho.group(1))
            else:
                sal= 0
                
                
        #find next page
        next_page = soup.find('a',{"id":"navigation-forward"})
        if next_page.get('href') =='#' or next_page.get('href') =='':
            next = False
        else:
            url = "https://internshala.com" +  next_page.get('href')
            

        
        if sal_must <= sal:
            i+=1
            #print(title +" ----> "+company +"   -----> "+location+"  ----> "+duration+"  ----->  "+salary+"  ##### "+str(sal))
            internship_detail[i] = [title,company,location,duration,salary,link]

            
print(str(i)+"  Internship Found")


internship = pd.DataFrame.from_dict(internship_detail, orient='index')
internship.columns = ['Job Title','Comapny Name','Location','Duration','Salary','Link']

#Export into CSV
internship.to_csv('internship.csv')
```
